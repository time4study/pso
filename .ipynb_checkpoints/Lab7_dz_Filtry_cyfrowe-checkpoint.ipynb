Spróbuj teraz funkcę *np.sinc*.

````python
from scipy.signal import freqz
N = 1000 #by byl filtr typu 1., musi byc nieparzysta liczba
h = sinc(linspace(-100,100,N))

plot(h)

f,H = freqz(h)

figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')
````

Wszystkie poprzednie filtry, to filtry dolno-przepustowe. Jak zrobić górno przepustowy? Spróbuj odwrócić znak co drugiego elementu sygnału.

````python
N = 1000
h = sinc(linspace(-N/10,N/10,N))

h[::2] *= -1
plot(h)

f,H = freqz(h)

figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')
````

Zróbmy połączenie dwóch filtrów (dolno i górno przepustowego). Do tego wystarczy zrobić sumę w dziedzinie częstotliowości, czyli również sumę w dziedzinie czasu.

````python
N = 1001
hlo = sinc(linspace(-N/10,N/10,N))
hhi = copy(hlo)

hhi[::2] *= -1

h = hhi + hlo

f,H = freqz(h)

figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')
````

## IIR - NOI

Filtry IIR są bardziej skomplikowane, ponieważ złe współczynniki mogą spowodować niestabilność układu (poprzez efekt *feedback*). Zaprojektuj filtr IIR za pomocą funkcji *scipy.signal.butter*. Użyj funkcji *scipy.signal.freqz* żeby obejrzeć jego widma, a *scipy.signal.lfilter* żeby zastosować filtr na sygnale.

Sensownymi wartościami filtru Butterwortha to $N=19$ i $W_n=0.5$.

````python
from scipy.signal import butter, lfilter #na zadanie mozna wykorzystac

b,a = butter(19, 0.5)

plot(b)
figure()
plot(a)

f,H = freqz(h)
figure()
plot(f,abs(H),'b')
# yscale('log')
twinx()
plot(f,angle(H),'g')
````

Spróbuj filtr eliptyczny używając funkcji *scipy.signal.ellip* z parametrami $N=10$, $rp=0.1$, $rs=10$ i $Wn=0.5$.

````python
from scipy.signal import ellip

b,a = ellip(10, 0.1, 10, 0.5)
f,H = freqz(b,a)
figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')
````

Policz odpowiedź impulsową tego filtru.

````python
i = zeros(100) #impuls
i[0] = 1

h = lfilter(b,a,i)
plot(h)
````

Sprawdź czy filtr jest stabilny. Filtr IIR jest stabilny jeśli wartości bezwględne pierwiastków jego współczynników mianownikowych są mniejsze od 1. Do policzenia pierwiastków można użyć funkcji *np.roots*.

````python
a[0]=30 #niestabilny
r = roots(a) #automatyznie wyliczy wszystkie pierwiastki

if all(abs(r)<1):
    print('Jest stabilny')
else:
    print('Jest niestabilny')
````

# Praca domowa

## 1. Filtr SOI

Zrób sygnał z trzema sygnałami: 1 kHz, 2kHz i 3 kHz. Zrób filtr który usunie składową 2 kHz.

````python
from scipy.signal import firwin, lfilter

T = 4 
Fs = 10000.0
t = arange(0,T,1/Fs)
f1 = 1000
f2 = 2000
f3 = 3000

y=sin(2*pi*f1*t)+sin(2*pi*f2*t)+sin(2*pi*f3*t)

figure()
plot(t,y)
xlim(0,0.01)

figure()

Y = fft.fft(y)
f = linspace(0,Fs,y.size)
plot(f,abs(Y))

cut_offs = [1900,2100]
numtaps = 31

h = firwin(numtaps, cut_offs, pass_zero=True, fs=Fs)

f,H = freqz(h,1)
f = (f/pi) * (Fs/2)
figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')

xf = lfilter(h, 1, y)

figure()

Y = fft.fft(xf)
f = linspace(0,Fs,Y.size)
plot(f,abs(Y))
````

## 2. Filtr NOI

Użyj tego samego sygnału co powyżej, ale zastosuj filtr NOI. Sprawdź czy filtr jest na pewno stabilny.

````python
from scipy.signal import firwin, lfilter

T = 4 
Fs = 10000.0
t = arange(0,T,1/Fs)
f1 = 1000
f2 = 2000
f3 = 3000

b,a = butter(19, 0.5)
y=sin(2*pi*f1*t)+sin(2*pi*f2*t)+sin(2*pi*f3*t)

figure()
plot(t,y)
xlim(0,0.01)

figure()

Y = fft.fft(y)
f = linspace(0,Fs,y.size)
plot(f,abs(Y))

f,H = freqz(b,a)
f = (f/pi) * (Fs/2)
figure()
plot(f,abs(H),'b')
twinx()
plot(f,angle(H),'g')

xf = lfilter(b, a, y)

figure()

Y = fft.fft(xf)
f = linspace(0,Fs,Y.size)
plot(f,abs(Y))
````
